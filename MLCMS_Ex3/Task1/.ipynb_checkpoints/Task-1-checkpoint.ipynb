# Part1 Principal component analysis

## Plot the data

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the data set
data = np.loadtxt('pca_dataset.txt')

# Step 2: Plot the data set
plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data Set')
plt.grid(True)
plt.show()
```

```python
# Step 3: Preprocess the data
data_centered = data - np.mean(data, axis=0)

# Step 4: Compute the SVD
U, s, Vt = np.linalg.svd(data_centered)

# Step 5: Determine the principal components
pc1 = Vt[0]
pc2 = Vt[1]

# Step 6: Calculate the energy in each component
total_energy = np.sum(s ** 2)
energy_pc1 = (s[0] ** 2) / total_energy
energy_pc2 = (s[1] ** 2) / total_energy

# Step 7: Plot the data set with principal components
plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
plt.quiver(np.mean(data[:, 0]), np.mean(data[:, 1]), pc1[0], pc1[1], color='r', scale=3)
plt.quiver(np.mean(data[:, 0]), np.mean(data[:, 1]), pc2[0], pc2[1], color='g', scale=3)

# Add labels to principal components
plt.text(pc1[0], pc1[1], 'PC1', color='r', fontsize=12, ha='center', va='center')
plt.text(pc2[0], pc2[1], 'PC2', color='g', fontsize=12, ha='center', va='center')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('PCA with Principal Components')
plt.grid(True)
plt.show()
```

## Energy contained in each of the two components

```python
print("Energy contained in each component:")
print("Component 1: {:.2%}".format(energy_pc1))
print("Component 2: {:.2%}".format(energy_pc2))
```

# Part2 Apply PCA to the image

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from sklearn.decomposition import PCA

# Load and preprocess the image
image_path = "PIXNIO-28860-1536x1152.jpeg"
image = Image.open(image_path).convert("L")  # Open image in grayscale
resized_image = image.resize((249, 185))  # Resize the image

# Display the resized image
plt.imshow(resized_image, cmap='gray')
plt.axis('off')
plt.show()
```

## Visualize reconstructions of the image

```python
data = np.array(resized_image).T  # Convert image to NumPy array and transpose

# Perform PCA on the image data
pca = PCA()
components = pca.fit_transform(data)

# Visualize reconstructions with different numbers of principal components
reconstructions = []
num_components = [len(pca.explained_variance_), 120, 50, 10]

for n in num_components:
    # Set singular values to zero after the nth component
    pca.components_[n:] = 0
    # Reconstruct the data using transformed components
    reconstructed_data = pca.inverse_transform(components)
    # Transpose the reconstructed data back to its original form
    reconstructed_image = reconstructed_data.T
    # Append the reconstructed image to the list
    reconstructions.append(reconstructed_image)

# Visualize the reconstructions
titles = ['All Components', '120 Components', '50 Components', '10 Components']

plt.figure(figsize=(12, 8))

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(reconstructions[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

```

## Analyse 

```python
# Calculate information loss and energy loss thresholds
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

threshold_index = np.argmax(cumulative_variance_ratio > 0.99)
visible_loss_index = np.argmax(explained_variance_ratio < 0.01)

visible_loss_num = visible_loss_index + 1
energy_loss_num = threshold_index + 1

print("Number where information loss becomes visible:", visible_loss_num)
print("Number where energy lost through truncation < 1%:", energy_loss_num)
```

# Part3 Analyse the trajectory data

## Visualize the path of the first two pedestrians in 2D space

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the data
data = np.loadtxt("data_DMAP_PCA_vadere.txt")

# Extract data for the first two pedestrians
pedestrian1 = data[:, 0:2]  # Assuming the x and y coordinates of pedestrian 1 are in the first two columns
pedestrian2 = data[:, 2:4]  # Assuming the x and y coordinates of pedestrian 2 are in the third and fourth columns

# Visualize the paths of the first two pedestrians
plt.plot(pedestrian1[:, 0], pedestrian1[:, 1], label="Pedestrian 1")
plt.plot(pedestrian2[:, 0], pedestrian2[:, 1], label="Pedestrian 2")
plt.xlabel("X position")
plt.ylabel("Y position")
plt.title("Paths of Pedestrians")
plt.legend()
plt.show()
```

## Check energy captured by the first two components

```python
# Perform Principal Component Analysis (PCA)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

# Calculate the percentage of energy captured by the first two components
energy_percentage = np.sum(pca.explained_variance_ratio_) * 100

print(f"Energy captured by the first two components: {energy_percentage:.2f}%")
```

## Check energy captured by the three components

```python
# Perform Principal Component Analysis (PCA)
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data)

# Calculate the percentage of energy captured by the three components
energy_percentage = np.sum(pca.explained_variance_ratio_) * 100

print(f"Energy captured by the three components: {energy_percentage:.2f}%")
```
