## For  = 1, numerically compute and visualize two orbits of the system (8)

```python
# Compute and plot trajectories
start_points = [(2, 0), (0.5, 0)]
t = np.linspace(0, 10, 1000)  # you may need to adjust the time range and number of points
alpha = 1

for start_point in start_points:
    trajectory = odeint(dynamic_system, start_point, t, args=(alpha,))
    plt.figure()
    plt.plot(trajectory[:, 0], trajectory[:, 1])
    plt.plot(start_point[0], start_point[1], 'bo')  # Plot the start point as a blue circle
    plt.title(f"Orbit for start point {start_point}")
    plt.show()
```
