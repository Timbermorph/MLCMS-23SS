```python
import numpy as np
import matplotlib.pyplot as plt
from utils import draw_phase_portrait
```

```python
#Unstable focus
alpha = 0.1
matrix_val = np.array([
    [alpha, alpha],
    [-1/4, 0]
])

draw_phase_portrait(matrix_val, alpha, "Unstable focus")
```

```python
#Unstable saddle
alpha = -0.5
matrix_val = np.array([
    [alpha, alpha],
    [-1/4, 0]
])

draw_phase_portrait(matrix_val, alpha, "Unstable saddle")
```

```python
#Unstable node
alpha = 3
matrix_val = np.array([
    [alpha, alpha],
    [-1/4, 0]
])

draw_phase_portrait(matrix_val, alpha, "Unstable node")
```

```python
#stable node
alpha = -1
matrix_val = np.array([
    [alpha, -0.5*alpha],
    [-1/4, 0]
])

draw_phase_portrait(matrix_val, alpha, "stable node")
```

```python
#stable focus
alpha = 0.1
matrix_val = np.array([
    [-alpha, alpha],
    [-1/4, 0]
])

draw_phase_portrait(matrix_val, alpha, "stable focus")
```
