Plot the mathematical function: alpha - x^2

```python
import numpy as np
import matplotlib.pyplot as plt

def dynamical_system(alpha, x):
    return alpha - x**2

# Set the alpha values
alpha_values = [-1, 0, 1]
x_values = np.linspace(-2, 2, 100)

# Plot the function for each alpha value
for alpha in alpha_values:
    y_values = dynamical_system(alpha, x_values)
    plt.plot(x_values, y_values, label=f'alpha = {alpha}')

plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = alpha - x^2 for different alpha values')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

Plot the phase portrait of the mathematical function

```python


import numpy as np
import matplotlib.pyplot as plt

# Define the dynamical system function
def dynamical_system(x, alpha):
    return alpha - 2*x**2 -3

# Set the value of alpha
alpha = 4

# Set the range of x values
x_values = np.linspace(-2, 2, 100)

# Set the range of t values
t_values = np.linspace(0, 5, 100)

# Create a meshgrid of x and t values
t, x = np.meshgrid(t_values, x_values)

# Compute the derivatives dx/dt
dxdt = dynamical_system(x, alpha)

# Plot the phase portrait
plt.figure(figsize=(8, 6))
plt.streamplot(t, x, np.ones_like(dxdt), dxdt, density=1, linewidth=1, arrowsize=0.5, color='b')
plt.xlabel('t')
plt.ylabel('x')
plt.title(f'alpha = {alpha}')
plt.grid()
plt.show()
```

Plot the bifurcation of the mathematical function: f(x) = alpha - x^2

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_bifurcation_plot(r_values):
    stable_x = []
    unstable_x = []
    for r in r_values:
        if r > 0:
            stable_x.append(np.sqrt(r))
            unstable_x.append(-np.sqrt(r))
        elif r < 0:
            unstable_x.append(0)
            stable_x.append(0)
        else:
            stable_x.append(0)
            unstable_x.append(0)
    return stable_x, unstable_x

# Parameters
r_values = np.linspace(-2, 5, 1000)  # Range of r values

# Compute bifurcation plot
stable_x, unstable_x = quadratic_bifurcation_plot(r_values)

# Plot the bifurcation plot
plt.figure(figsize=(10, 6))
plt.plot(r_values, stable_x, 'b-', markersize=0.1, label='Stable')
plt.plot(r_values, unstable_x, 'r-', markersize=0.1, label='Unstable')
plt.xlabel('alpha')
plt.ylabel('x')
plt.title('Bifurcation Plot:')
plt.legend()
plt.show()
```
